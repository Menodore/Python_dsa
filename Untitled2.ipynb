```python
import opendatasets as od
import pandas as pd

od.download("https://www.kaggle.com/datasets/akshaydattatraykhare/diabetes-dataset")
```

```python
df=pd.read_csv("diabetes-dataset/diabetes.csv")
df.head(5)
df.dropna()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter the data
df_diabetic = df[df['Outcome'] == 1]
df_nondiabetic = df[df['Outcome'] == 0]

# Plotting
plt.figure(figsize=(8, 6))

# Scatter plot for non-diabetic cases
plt.scatter(df_nondiabetic['Glucose'], df_nondiabetic['Insulin'], label='Non-diabetic', color='b', alpha=0.5)

# Scatter plot for diabetic cases
plt.scatter(df_diabetic['Glucose'], df_diabetic['Insulin'], label='Diabetic', color='r', alpha=0.5)

plt.xlabel('Glucose Level')
plt.ylabel('Insulin Level')
plt.title('Correlation between Insulin and Glucoses')
plt.legend()
plt.show()
```

```python


plt.figure(figsize=(8, 6))
plt.scatter(df['BMI'], df['BloodPressure'], c=df['Age'], cmap='viridis')
plt.xlabel('BMI')
plt.ylabel('BP')
plt.title('BP v/s BMI by Age')
_ = plt.colorbar(label='Age')
```

```python
import seaborn as sns

# Plot the distribution of the Outcome variable
plt.figure(figsize=(8, 6))
sns.countplot(x='Outcome', data=df)
plt.title('Distribution of Diabetes Outcome')
plt.xlabel('Outcome')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))
sns.heatmap(df.corr(),annot = True,  cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Diabetes Dataset Features')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the data for diabetic cases with at least one pregnancy
df_diabetic_pregnant = df[(df['Outcome'] == 1) & (df['Pregnancies'] >= 1)]

plt.figure(figsize=(8, 6))
sns.histplot(df_diabetic_pregnant['Age'], kde=True, bins=30)
plt.title('Hisplot of Age Distribution for Diabetic Individuals with at Least One Pregnancy')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```
